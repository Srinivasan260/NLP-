```python
##drop the null values

df=df.dropna()   ##it removes the null values
```

```python
##taking the dependent feature and dependent features

X=df.drop('label',axis=1)

```

```python
##get the dependent features
y=df['label']
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
##vocubalary size
voc_size=10000
```

```python
##one hot encoding

message=X.copy()

message['title'][1]
```

```python
message
```

```python
message.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
mesaages=message.head(20)
```

```python
corpus = []
for i in range(0 , len(message)):
    review = re.sub('[^a-zA-Z]', '  ', message['title'][i])
    review = review.lower()
    review = review.split()
    
    review =[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus]
onehot_repr
```

```python
corpus[1]
```

```python
##embedding representation

sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[1]
```

```python
#creatung the model

embedding_vector_features=40 ##each and every word in previous array is converted into a vector
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid')) ##whenever my output is 1 or 0 use sigmoid activation
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
X_final=np.array(embedded_docs) ##just converting the lists into array
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size=0.33,random_state=42)
```

```python
##nodel_training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

Performance metrics and accuracy

```python
y_pred = model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
